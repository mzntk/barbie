```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('/Users/majazontek/Desktop/UCZELNIA/2 rok/wizualizacja danych/projekt/do wyslania do 23.12.25/pozostałe (do 10.01.2026)/MAT_history_monthly.csv', sep=';')

df.drop(1, inplace=True)
df.drop(0, inplace=True)
df = df.rename(columns={'Price': 'Date'})

df['Date'] = pd.to_datetime(df['Date'])
df['Close'] = pd.to_numeric(df['Close'])
df = df.dropna(subset=['Close'])

df = df[df['Date'].dt.year >= 2004]

plt.figure(figsize=(16, 9), facecolor='lightgray')
plt.plot(df['Date'], df['Close'], color='blue', label='Closing price (Close)')

plt.title('Historical MAT Share Price (since 2004)')
plt.xlabel('Date')
plt.ylabel('Price ($)')
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()

plt.axvline(x=pd.to_datetime('2014'), color='red', linestyle='--')
plt.axvline(x=pd.to_datetime('2023'), color='red', linestyle='--')

l = []
for i in range(0, 5):
    l.append(i*5 + 2005)


l.append(2004) # Początek wykresu
l.append(2014) # Wydarzenie 1
l.append(2023) # Wydarzenie 2

l = sorted(list(set(l)))

l_str = [str(x) for x in l]
g = [pd.to_datetime(str(x)) for x in l]

plt.xticks(g, l_str, rotation=45, ha='right')

plt.tight_layout()
plt.savefig('MAT-gielda_2004.png')
plt.show()
```
