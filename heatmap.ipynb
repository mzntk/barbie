```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import calendar

df = pd.read_csv('/Users/majazontek/Desktop/UCZELNIA/2 rok/wizualizacja danych/projekt/do wyslania do 23.12.25/pozostałe (do 10.01.2026)/wyszukiwanie-barbie-2015-2024.csv', header=1)

df['date'] = pd.to_datetime(df['Miesiąc'])
df['Year'] = df['date'].dt.year
df['Month'] = df['date'].dt.month

heatmap_data = df.pivot(index='Year', columns='Month', values='Barbie: (Cały świat)')

plt.figure(figsize=(12, 6)) #szerszy format

ax = sns.heatmap(heatmap_data,
            annot=True,
            fmt=".0f",      #miejsca po przecinku (usuniecie)
            cmap="RdPu",
            linewidths=0.5,
            linecolor='white',
            cbar_kws={'label': 'Popularity index (0-100)'})

ax.set_xticklabels([calendar.month_abbr[i] for i in range(1, 13)])
plt.yticks(rotation=0)      #poziome etykiety na osi y

plt.title('Search popularity for “Barbie” (2015-2024)\n'
          'Visible “Barbenheimer” effect in July 2023',
          fontsize=14, loc='left', pad=20)

plt.xlabel('Month', fontsize=12)
plt.ylabel('Year', fontsize=12)

sns.despine(top=True, right=True, left=True, bottom=True) #usuniecie ramek

plt.tight_layout()
plt.savefig('heatmap.png')
plt.show()
```
